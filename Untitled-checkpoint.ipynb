```python
data.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

```python
# convert  categorical into numerical variable
data['Gender'].value_counts()
```

```python
data['Geography'].value_counts()
```

```python
# encode categorical variable

encode_gender=LabelEncoder()
data['Gender']=encode_gender.fit_transform(data['Gender'])
data
```

```python
# encode geography column using one hot encoding
from sklearn.preprocessing import OneHotEncoder
ohe_geo=OneHotEncoder()
geo_encoder=ohe_geo.fit_transform(data[['Geography']]).toarray()
geo_encoder
```

```python
ohe_geo.get_feature_names_out(['Geography'])
```

```python
# Convert the result to a DataFrame with proper column names (optional)
geo_encoded_df = pd.DataFrame(geo_encoder, columns=ohe_geo.get_feature_names_out(['Geography']))

geo_encoded_df
```

```python
# combine the dataset

data=pd.concat([data.drop('Geography',axis=1),geo_encoded_df],axis=1)
data.head()

#import pandas as pd

# Assuming 'data' is your DataFrame
#geo_encoded_df = pd.get_dummies(data['Geography'], prefix='Geography', drop_first=True)

#geo_encoded_df
```

```python
# save encoder and scaler into pickle file
with open('label-encode_gender.pkl','wb') as file:
    pickle.dump(encode_gender,file)

with open('onehot_encode_geo.pkl','wb') as file:
    pickle.dump(ohe_geo,file)
```

```python
# split the dataset
X=data.drop('Exited',axis=1)
y=data['Exited']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
# scale the dataset
standard_scaler=StandardScaler()
X_train=standard_scaler.fit_transform(X_train)
X_test=standard_scaler.transform(X_test)
```

```python
with open('scaler.pkl','wb') as file:
    pickle.dump(standard_scaler,file)
```

```python
# ANN implementation
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import datetime
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
```

```python
!pip install tensorflow
```

```python
X_train.shape[1],
```

```python
# build ANN model
model=Sequential([
    Dense(64,activation='relu',input_shape=(X_train.shape[1],)),         # hidden layer1
    Dense(32,activation='relu'),      #hidden layer 2
    Dense(1,activation='sigmoid')    # output layer
])
```

```python
model.summary()
```

```python
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
loss=tensorflow.keras.losses.BinaryCrossentropy()
loss
```

```python
## compile the model
model.compile(optimizer=opt,loss="binary_crossentropy",metrics=['accuracy'])
```

```python
## Set up the Tensorboard
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard

log_dir="logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
## Set up Early Stopping
early_stopping_callback=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
### Train the model
history=model.fit(
    X_train,y_train,validation_data=(X_test,y_test),epochs=100,
    callbacks=[tensorflow_callback,early_stopping_callback]
)
```

```python
model.save('model.h5')
```
